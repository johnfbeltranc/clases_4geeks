2. cual es la probabilidad de obtener un numero primo o un numero par en un lanzamiento de dados

```python
s = {1, 2, 3, 4, 5, 6}   # espacio muestral
e1 = {2, 3, 5}
e2 = {2, 4, 6}
e = {2, 3, 4, 5, 6}    # union
e = set.union(e1, e2)   # union de los eventos
c_n = len(set.intersection(e, s))   # cardinalidad de la intersecci贸n del evento y el espacio muestra,cardinalidad de numerador
c_d = len(s)    # cardinalidad del espacio muestra,cardinalidad del denominador
Fraction(c_n,c_d) # la probabilidad de que ocurra el evento e, entre, el espacio muestral s
```

3. cual es la probabilidad de obtener un numero par y primo en un lanzamiento de dados

```python
# cuando en el enunciado es "y" entonces es interseccion cuando es "o" entonces es union
s = {1, 2, 3, 4, 5, 6} # espacio muestral
e1 = {2, 4, 6}
e2 = {2, 3, 5}
e = set.intersection(e1, e2) # intersecci贸n de los eventos
c_n = len(set.intersection(e, s)) # cardinalidad de la intersecci贸n del evento y el espacio muestra
c_d = len(s) # cardinalidad del espacio muestra
Fraction(c_n,c_d)
```

```python
def Pr(event:set,space:set) -> Fraction:
    return Fraction(len(set.intersection(event, space)), len(space))
    
```

4. el metodo de montecarlo nos permite aproximar una probabilidad en base a la repeticion continua de un evento. cual es la probabilidad de obtener un numero par en un lanzamiento de dados

```python
events = [2, 4, 6]
space = [1, 2, 3, 4, 5, 6]
frec = []
trials = list(range(1, 1001)) # 1000 trials intentos
for trial in trials:
    total = 0
    for i in range(trial):
        if random.choice(space) in events: # coge un numero a la zar de space y si esta el mismo en events es un exito
            total = total + 1 # si el numero elegido aleatoriamente esta en los eventos, entonces sumamos 1
    frec.append(total/trial)   # divide total entre trial numero de intentos 
```

```python
plt.plot(trials, frec)
plt.xlabel("trials")
plt.ylabel("relative frequency")
plt.axhline(y=0.5,color="tomato")
plt.grid(True)
plt.show()
```
