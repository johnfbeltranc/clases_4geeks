```python
# step 0
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
step 1
url = 'https://en.wikipedia.org/wiki/List_of_U.S._states_by_intentional_homicide_rate' # URL de la página que quiero scrapear
headers = {"User-Agent":"MywebScrapper/1.0"} # este es mi user-agent
response = requests.get(url, headers=headers) # hago la petición a la página
soup = BeautifulSoup(response.content, 'html.parser') # parseo el contenido HTML
soup
```

```python
my_tables = soup.find_all('table', class_='wikitable') # busco todas las tablas
print(f"Number of tables:{len(my_tables)}") # cuento cuántas tablas hay
```

```python
my_tables[0]
```

```python
rows = my_tables[0].find_all('tr') # busco todas las filas de la primera tabla
rows
```

```python
my_tables[0]
rows = my_tables[0].find_all('tr') # busco todas las filas de la primera tabla
my_headers = rows[0] # la primera fila es la que contiene los headers
header_cells = my_headers.find_all('th') # busco todas las celdas de encabezado
header_data = [] # lista vacía para guardar los headers
header_data = [cell.get_text(strip=True) for cell in header_cells] # obtengo el texto de cada celda
header_data 
```

```python
content_cells = rows[1:]
content_data = []
for i, data_cells in enumerate(content_cells):
    cells = data_cells.find_all('td') # busco todas las celdas de datos
    row_data = [cell.get_text(strip=True) for cell in cells] # obtengo el texto de cada celda
    content_data.append(row_data) # agrego los datos de la fila a la lista
content_data    
```

```python
import pandas as pd

# Crear DataFrame usando los headers como columnas
df = pd.DataFrame(content_data, columns=header_data)

# Ver primeras filas
print(df.head(10))
```

```python
df["2018"] = df[2018].as
```
