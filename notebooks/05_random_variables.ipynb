el modelo bernoulli se aplica a un random variable que puede asumir solo dos valores. para este modelo tenemos:
- $E(X) = p$
- $V(X=pq$)

donde p ,es la probabilidad de exito y q es la probabilidad de fracaso. $E(X)$ significa la esperanza de la variable aleatoria X y $V(X)$ significa la varianza de la variable aleatoria X

El modelo binomial puede considerarse como la suma de n variables bernulli independientes una de otras donde se cumpla lo siguiente:
1. debe haber un numero fijo de pruebas repetidas estadisticamente independientes.
2. cada prueba debe ser una variable bernoulli, es decir puede resultar en un exito o un fracaso
3. todas las pruebas deben tener identicas probabilidades de exito

El modelo binomial es discreto y tiene lo siguiente:
- $E(X)=np$
- $V(X)=npq$

Donde n es el numero de ensayos o pruebas, p es la probabilidad de exito y q es la probabilidad de fracaso.

Una moneda es lanzada 10 veces. Cual es la probabilidad de obtener 8  o mas caras?

```python
n = 10 # número de lanzamientos
p = 0.5 # probabilidad de éxito (cara)
# probabilidad de obtener 8 o más caras en 10 lanzamientos de una moneda p(X >= 8)
prob = 1 - binom.cdf(7, n, p) # significa que hay una probabilidad de 1 - prob de que se obtengan más de 7 éxitos en 10 ensayos
prob
```

Se sabe que el 90% de los trabajadores de una ciudad industrial son mienbros de un sindicato. En un estudio de situacion laboral de la ciudad un economista escoge al azar una muesttra de 15 obreros para entrevistas. Cuales la probabilidad de que la muestra contenga 12 0 menos mienbros del sindicato.

```python
p = 0.9 # probabilidad de éxito (miembros del sindicato)
n = 15 # número de trabajadores
# p(X <= 12) = P(X=0) + P(X=1) + ... + P(X=12)
prob = binom.cdf(12, n, p) # significa que hay una prob
prob
```

muchos experimentos no ucurren como un resultado definido de prueba sino como puntos de tiempo., espacios o volumen a la zar. el modelo poisson es adecuado en estos casos si: 
1. el numero de ocurrencias del hecho es independiente de una unidad
2. el valor esperado de la variable es proporcional al tamano de la unidad especificada
3. la probabilidad de mas de una ocurrencia del hecho es una unidad especificada muy pequena es despreciable

para este modelo tenemos:

- $E(X)=\lambda$
- $V(X)=\lambda$

Donde $lambda$ es igual a el numero de ocurrencias por unidad especificada. la unidad poisson es discreta

En un callcenter ocurren 4 llamadas por minuto ($\lambda$)=4. cual es la posibilidad de que no ocurran mas de 2 llamadas durante 2 minutos siguientes?

```python
n = 2 # número de eventos
plambda = 8 # tasa promedio de ocurrencia
# p(X<=2) es decir p(X=0) + p(X=1) + p(X=2)
prob = poisson.cdf(n, plambda  ) # significa que hay una probabilidad de prob de que ocurran 2 o menos eventos en un intervalo de tiempo dado
prob
```

```python
# p(x>=4) = p(X=4) + p(X=5) + p(X=6) + ...
#p(X>=4) = 1 - p(X<=3) = 1- p(X=0) - p(X=1) - p(X=2) - p(X=3)
n = 3 # número de eventos
plambda = 8 # tasa promedio de ocurrencia
prob = 1 - poisson.cdf(n, plambda) # significa que hay una probabilidad de prob de que ocurran 4 o más eventos en un intervalo de tiempo dado
prob
```

un fabricante de tejido de lana afirma que el promedio de defectos en sus productos es 1 por 2 yardas cuadradas. se estrajo una yarda cuadrada al azar y tenia 3 defectos. cual es la probabilidad de obtener 3 o mas defectos en 3 o mas yardas cuadradas si la afirmacion del fabricante fuera balida

```python
plambda = 0.5 # tasa promedio de ocurrencia se divide 2 yardas por un defecto
n = 3 # número de eventos
# p(X>=3) = p(X=3) + p(X=4) + p(X=5) + ...
# p(X>=3) = 1 - p(X<=2) = 1 - p(X=0) - p(X=1) - p(X=2)
prob = 1 - poisson.cdf(n, plambda) # probabilidad de que ocurran 3 o más eventos in un intervalo de tiempo dado
prob
```

El modelo de probabilidad mas frecuentemente usado en el analisis economico y comercial es la distribucion normal, que puede expresarse en las formas general y estandarizada. se dice que una variable aleatoria X tiene una distribucion normal si es continua, si esta en el rango del menos infinito al infinito y si puede expresarse por su media 

```python
x = np.linspace(-5, 5, 100) # rango de valores
pdf = norm.pdf(x) # función de densidad de probabilidad
plt.plot(x, pdf, c='blue') # gráfica de la distribución normal
plt.title('Distribución Normal') # título de la gráfica
plt.xlabel('value') # etiqueta del eje x
plt.ylabel('Density') # etiqueta del eje y
plt.grid(True) # activar la cuadrícula
plt.show()
```

```python
x = np.linspace(-5, 5, 100) # rango de valores
cdf = norm.cdf(x) # función de distribución acumulativa
plt.plot(x, cdf, c='blue') # gráfica de la distribución normal
plt.title('Distribución Normal') # título de la gráfica
plt.xlabel('value') # etiqueta del eje x
plt.ylabel('probability') # etiqueta del eje y
plt.grid(True) # activar la cuadrícula
plt.show()
```

```python
n = 10 # número de ensayos
p = 0.5 # probabilidad de éxito
x = range(n+1) # número de ensayos
pmf = binom.pmf(x, n, p) # función de masa de probabilidad
plt.stem(x, pmf) # gráfica de la distribución binomial
plt.grid(True) # activar la cuadrícula
plt.show() # mostrar la gráfica
```
