```python
import numpy as np 
import matplotlib.pyplot as plt
import sympy as sy
```

```python
def line_function(x:float)->float: 
  return x+2 # Esta función lineal toma un número y le suma 2 
 #Si X es 0, y = 2, si X es 3 y seria 5. La notación matemática f(x) = x+2
```

```python
x = np.linspace(-5,5,100) # Crea un vector de 100 puntos equidistantes entre -5 y 5
plt.axhline(y=0, color="r") # Dibuja la línea horizontal en y=0
plt.axvline(x=0, color="r") # Dibuja la línea vertical en x=0
plt.grid(True) # Activa la cuadrícula
plt.plot(x, line_function(x), label="linefunction") # Dibuja la función lineal
plt.show() # Muestra la gráfica
```

```python
def f(x:float)->float: # Esta función lineal toma un número y le multiplica por 2 y le suma 2
  return 2 * x + 2
```

```python
x = np.linspace(-5,5,100) # Crea un vector de 100 puntos equidistantes entre -5 y 5
plt.axhline(y=0, color="r") # Dibuja la línea horizontal en y=0
plt.axvline(x=0, color="r") # Dibuja la línea vertical en x=0
plt.grid(True) # Activa la cuadrícula
plt.plot(x, f(x), label="f") #Dibuja la función f(x) respecto a x en un gráfico de líneas y le asigna la etiqueta "f"
plt.show() # Muestra la gráfica
```

```python
def g(x:float)->float: # Esta función cuadrática toma un número y le aplica la operación x^2 + 1
  return x * x + 1
```

```python
x = np.linspace(-5,5,100) # Crea un vector de 100 puntos equidistantes entre -5 y 5
plt.axhline(y=0, color="g") # Dibuja la línea horizontal en y=0
plt.axvline(x=0, color="g") # Dibuja la línea vertical en x=0
plt.grid(True) # Activa la cuadrícula
plt.plot(x, g(x), label="g") # Dibuja la función g(x) respecto a x en un gráfico de líneas y le asigna la etiqueta "g"
plt.plot(0, g(0), "ro" ) # Dibuja un punto rojo en (0, g(0))
plt.plot(2, g(2), "ro") # Dibuja un punto rojo en (2, g(2))
plt.show()
```

```python
def k(x:float)->float: #funcion cuadratica con termino lineal
  return x * x - 2 * x + 1

def dk(x:float)->float: # Esta función calcula la derivada de k(x)
  h = 0.000001 # Un valor pequeño para aproximar la derivada
  return (k(x+h)-k(x))/h # Retorna la pendiente de la tangente en x
```

```python
x = np.linspace(-5,5,100) # Crea un vector de 100 puntos equidistantes entre -5 y 5
plt.axhline(y=0, color="g") # Dibuja la línea horizontal en y=0
plt.axvline(x=0, color="g") # Dibuja la línea vertical en x=0
plt.grid(True) # Activa la cuadrícula
plt.plot(x, k(x), label="k(x)") # Dibuja la función k(x) respecto a x en un gráfico de líneas y le asigna la etiqueta "k(x)"
plt.plot(x, dk(x), label="dk(x)") # Dibuja la función dk(x) respecto a x en un gráfico de líneas y le asigna la etiqueta "dk(x)"
plt.show()
```

```python
def square(x:float)->float: # Esta función calcula el cuadrado de un número
  return x ** 2 #significa potencia (x*x)
```

```python
plt.plot(x, square(x)) # Dibuja la función cuadrática respecto a x en un gráfico de líneas
plt.axhline(y=0, color="r") # Dibuja la línea horizontal en y=0
plt.axvline(x=0, color="r") # Dibuja la línea vertical en x=0
plt.grid(True) # Activa la cuadrícula
plt.title("Gráfica de la función cuadrática en el intervalo de 0 hasta 4") # Establece el título del gráfico
plt.show()
```

```python
x = np.linspace(-10, 10, 1000) # Crea un vector de 1000 puntos equidistantes entre -10 y 10

def f(x:float)->float: # Esta función calcula el seno de un número
  return np.sin(x)

plt.plot(x, f(x)) # Dibuja la función seno respecto a x en un gráfico de líneas
plt.axhline(y=0, color="r") # Dibuja la línea horizontal en y=0
plt.axvline(x=0, color="r") # Dibuja la línea vertical en x=0
plt.grid(True) # Activa la cuadrícula
plt.title("Gráfica de la función seno en el intervalo de -3 hasta 3")
plt.show()
```

```python
x = np.linspace(-3, 3, 1000) # Crea un vector de 1000 puntos equidistantes entre -3 y 3

def g(x:float)->float: # Esta función calcula la potencia de 2 de un número
  return np.power(2, x)

plt.plot(x, g(x)) # Dibuja la función exponencial respecto a x en un gráfico de líneas
plt.axhline(y=0, color="r") # Dibuja la línea horizontal en y=0
plt.axvline(x=0, color="r") # Dibuja la línea vertical en x=0
plt.grid(True) # Activa la cuadrícula
plt.title("Gráfica de la función exponencial en el intervalo de -3 hasta 3")
plt.show()
```
