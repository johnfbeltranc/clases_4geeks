Ha quedado bien establecido por los cientificos médicos que si el contenido medio de nicotina de los cigarrillos es de 25mg o más, es probable que el fumador contraiga cancer. Si el contenido de nicotina medio es menor de 25mg, el usuario se encunetra relativamente seguro.

Para establecer la hipótesis nuestro interés es muestrear una cantidad de cigarrillos y ver si el promedio de esta muestra es mayor a 25.

H0: $μ$ >= 25  
H1: $μ$ < 25

Un proceso industrial usado por una fábrica durante algunos años da una producción media de 100 unidades por hora, con una desviación estandar de 8 unidades. Acaba de ponerse en el mercado una nueva máquina para producir el mismo prodcuto. Aunque es muy costosa comparada con la que actualmente tenemos, su adopción sería muy lucrativa si su producción media fuera mayor a 150 unidades por hora. Suponga que la nueva máquina ha sido probada 35 horas y se ha encontrado que el promedio de producción es de 160 unidades por hora. Explicar su decisón respecto a la compra de la nueva máquina.

H0 = $μ$ <= 150  
H1 = $μ$ > 150

$\bar{x}$ = 160  
$σ$ = 8  
n = 35  

z = $\frac{\bar{x} - μ0}{σ/\sqrt{n}}$  



```python
#La hipotesis alterna es lo que quiero probar
n = 35
sample_mean = 160 # Media de la muestra
poblational_mean = 150 # Media poblacional
poblational_std = 8 # Desviación estándar poblacional
alfa = 0.05 # Estamos dispuestos a aceptar un error de 5% (1-alfa es lo que estamos dispuestos a estar en lo correcto)

z_score = (sample_mean - poblational_mean) / (poblational_std / np.sqrt(n)) # Calculamos el estadístico z
z_score #Es la cantidad de desviación standar alejado de la media

p_value = 1 - stats.norm.cdf(abs(z_score)) # Calculamos el p-value
p_value
# Si pvalue < alfa -> queda H1 sino H0
# np.float64(7.061018436615996e-14) < 0.05 -> nos quedamos con H1
```

La distancia media que debe recorrer un automóvil de cierta marca para detenerse cuando marcha a 30 millas por hora es 65 pies. El departamento de ingeniería de la compañia ha diseñado un nuevo sistema de frenos considerado como más eficaz. Para probar esta invención el nuevo sistema de frenos es instalado en 64 autos; las pruebas demuestran que la distancia media para detener un automovil a una velocidad de 30 millas por hora es 63.5 pies, con una desviación estandar de 4 pies. Esta reducción de distancia ¿Demuestra que el sistema de freno es más eficaz que el antiguo?

H0: $μ$ >= 65  #Esto es lo contrario  
H1: $μ$ < 65 #Es lo que quiero probar  

Nivel de significancia: $\alpha$ = 0.01  





```python
n = 64
sample_mean = 63.5
poblational_mean = 65
poblational_std = 4
alfa = 0.01 #Estamos dispuestos a aceptar un error de 1% (1-alfa es lo que estamos dispuesto a estar en lo correcto)

z_score = (sample_mean - poblational_mean) / (poblational_std / np.sqrt(n)) # Calculamos el estadístico z
z_score #Es la cantidad de desviación standar alejado de la media

p_value = 1 - stats.norm.cdf(abs(z_score))
p_value

# Si pvalue < alfa -> queda H1 sino H0
# np.float64(0.0013498980316301035) < 0.01 -> nos quedamos con H1
```

Un fabricante de cera para el piso tiene vendedores en toda la nación. En el pasado sus vendedores ganaban una comisión promedio de 600 dólares mensuales, con una desviación estandar de 80 dólares. Recientemente nuevas marcas de cera para piso han entrado al mercado, y este factor tiende a reducir el volumen de ventas por vendedor, y por tanto el ingreso de los vendedores. Suponga que se extrae al azar una muestra de 100 comisiones y se encuentra que el promedio es de 585 dólares. La gerencia de la empresa est ansiosa por descubrir si ha habido un efecto neto en las comisiones por la entrada de esta nueva marca.

H0 :  $\mu$ = 600  
H1 : $\mu \neq$ 600 #Lo que ha cambiado  

Nivel de signifcancia: $α$ = 0.05


```python
n = 100
sample_mean = 585
poblational_mean = 600
poblational_std = 80
alfa = 0.05 #Estamos dispuestos a aceptar un error de 5% (1-alfa es lo que estamos dispuesto a estar en lo correcto)

z_score = (sample_mean - poblational_mean) / (poblational_std / np.sqrt(n))
z_score #Es la cantidad de desviación standar alejado de la media

p_value = 1 - stats.norm.cdf(abs(z_score))
p_value

# Si pvalue < alfa -> queda H1 sino H0
# 0.025 = 0.05/2 por que el H1 no es igual
# np.float64(0.030396361765261393) > 0.025 -> nos quedamos con H0
```

A farmer decides to test three different types of fertilizers to determine if one is superior in terms of corn production. The farmer plants corn on 15 identical plots and uses all three fertilizers (5 plots for each type). At the end of the season, he measures the corn yield (in kg) of each plot, with the following result:

|Fertilizer 1|	Fertilizer 2|	Fertilizer 3|
|-|-|-|
|20|22|24|
|21|21|23|
|20|23|22|
|19|22|23|
|20|21|24|

With this data, he seeks to answer the following question: Is there a significant difference in average corn yield between the three types of fertilizers?

To help you, follow the points below:

State the hypothesis: null and alternative hypothesis.
Perform the ANOVA test.
Analyze the conclusions.
If one fertilizer is better than another, how can we know it?

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd # # Test de Tukey HSD para comparaciones múltiples después de ANOVA, Comparaciones múltiples entre grupos
data = [20,21,20,19,20,22,21,23,22,21,24,23,22,23,24] # Creamos los datos de ejemplo
labels = ['F1','F1','F1','F1','F1','F2','F2','F2','F2','F2','F3','F3','F3','F3','F3'] # Creamos las etiquetas
result = pairwise_tukeyhsd(data,labels,alpha=0.05) # Realizamos la prueba de Tukey
print(result)
#alfa = 0.05
#pvalue < alfa -> se cumple la hipotesis alterna (Hay diferencias)
```
